```python
#2.1.1

import torch

x=torch.arange(12) 
print("torch.arange(12) ==>\n",x,"\n")

print("torch.arange(12).shape: ",x.shape) #the length along each axis

print("torch.arange(12).numel(): ",x.numel()) #the total number of elements in a tensor

print()
A=x.reshape([3,4]) #轉換成3*4的matirx
print("torch.arange(12).reshape([3,4])==>\n",A,"\n")

A=x.reshape([-1,4])
print("torch.arange(12).reshape([-1,4])==>\n",A,"\n")

a=torch.arange(20)
A=a.reshape(2,2,5)
print("torch.arange(20).reshape(2,2,5)==>\n",A,"\n")

A=torch.zeros([2,3,4])
print("torch.zeros([2,3,4])==>\n",A,"\n")

A=torch.ones([2,3,4])
print("torch.ones([2,3,4])==>\n",A,"\n")

A=torch.randn(3,4)
print("torch.randn(3,4)==>\n",A,"\n")

A=torch.tensor([[2, 1, 4, 3], [1, 2, 3, 4], [4, 3, 2, 1]])
print("torch.tensor([[2, 1, 4, 3], [1, 2, 3, 4], [4, 3, 2, 1]])==>\n",A,"\n")

```

```python
#2.1.2

x = torch.tensor([1.0, 2, 4, 8])
y = torch.tensor([2, 2, 2, 2])

print("x ==>\n",x,"\n")
print("y ==>\n",y,"\n")
print("x+y ==>\n",x+y,"\n")
print("x-y ==>\n",x-y,"\n")
print("x*y ==>\n",x*y,"\n")
print("x/y ==>\n",x/y,"\n")
print("x**y ==>\n",x**y,"\n")
print("e^x ==>\n",torch.exp(x),"\n")


X = torch.arange(12, dtype=torch.float32).reshape((3, 4))
Y = torch.tensor([[2.0, 1, 4, 3], [1, 2, 3, 4], [4, 3, 2, 1]])
print("X ==>\n",X,"\n")
print("Y ==>\n",Y,"\n")

print("以row把X跟Y作合併 ==>\n",torch.cat((X, Y), dim=0),"\n")
print("以col把X跟Y作合併 ==>\n",torch.cat((X, Y), dim=1),"\n")
print("X==Y ==>\n",X==Y,"\n")
print("X的element總和 ==>\n",X.sum(),"\n")
```

```python
#2.1.5

X=torch.arange(12).reshape(3,4)
Y=torch.tensor([[2,1,4,3],[0,1,2,3],[4,5,6,7]])
print("X ==>\n",X,"\n")
print("Y ==>\n",Y,"\n")

before = id(Y) #Y的位址
Y = Y + X
print("Y=Y+X後, Y的新位址跟舊位址一不一樣? ==>\n",id(Y) == before,"\n")

Z = torch.zeros_like(Y)
print("Z ==>\n",Z,"\n")
print("id(Z): ", id(Z),"\n")

Z[:] = X + Y
print("Z ==>\n",Z,"\n")
print("id(Z): ", id(Z),"\n")

before = id(X)
X += Y
print("X+=Y後, X的新位址跟舊位址一不一樣? ==>\n",id(X) == before,"\n")
```

```python
#2.1.6

#格式轉換

X=torch.arange(12).reshape(3,4)
print("torch.arange(12).reshape(3,4) ==>\n",X,"\n")

A = X.numpy()
B = torch.tensor(A)

print("torch.arange(12).reshape(3,4).numpy() ==>\n",A,"\n")
print("type :\n",type(A),"\n")
print("torch.tensor(torch.arange(12).reshape(3,4).numpy()) ==>\n",B,"\n")
print("type :\n",type(B),"\n")

a = torch.tensor([3.5])
print("torch.tensor([3.5]) ==>\n",a,"\n")
print("torch.tensor([3.5]).item() ==>\n",a.item(),"\n")
print("float(torch.tensor([3.5])) ==>\n",float(a),"\n")
print("int(torch.tensor([3.5])) ==>\n",int(a),"\n")

a = torch.tensor([[3.5,5,5.1]])
print("torch.tensor([[3.5,5,5.1]]) ==>\n",a,"\n")
print("torch.tensor([[3.5,5,5.1]])[0,2] ==>\n",a[0,2],"\n")
print("torch.tensor([[3.5,5,5.1]])[0,2].item() ==>\n",a[0,2].item(),"\n")
print("float(torch.tensor([[3.5,5,5.1]])[0,2]) ==>\n",float(a[0,2]),"\n")
print("int(torch.tensor([[3.5,5,5.1]])[0,2]) ==>\n",int(a[0,2]),"\n")
```
